```python
import pandas as pd
import requests

# Load dataset (replace 'your_dataset.csv' with actual file)
df = pd.read_csv('superstore.csv')

# Print column names to check if 'city' exists
print("üîπ Column Names:", df.columns)

# Standardize column names (convert to lowercase and remove spaces)
df.columns = df.columns.str.lower().str.strip()

# Check for variations of 'city'
possible_city_columns = [col for col in df.columns if 'city' in col]
print("üîπ Possible city column:", possible_city_columns)

# Rename the correct city column to 'city' if needed
if possible_city_columns:
    df.rename(columns={possible_city_columns[0]: 'city'}, inplace=True)
else:
    raise KeyError("‚ùå No 'city' column found in dataset!")

# API Key & Base URL
API_KEY = "f605e05737a1c635a85d88f37fb98d51"
BASE_URL = "http://api.openweathermap.org/data/2.5/weather"

# Function to get weather details
def get_weather(city):
    try:
        url = f"{BASE_URL}?q={city}&appid={API_KEY}&units=metric"
        response = requests.get(url)
        data = response.json()

        if response.status_code == 200:
            return data['weather'][0]['description']  # Returns weather description (e.g., "clear sky")
        else:
            return "Unknown"
    except:
        return "Error"

# Add a new 'weather' column
df['weather'] = df['city'].apply(lambda x: get_weather(str(x)))

# Save the updated dataset
df.to_csv('updated_dataset.csv', index=False)

print("‚úÖ Weather column added successfully!")
```
